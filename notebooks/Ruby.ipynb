```ruby
require "active_record"
require "sqlite3"
require "action_view"

db_path = File.expand_path("database/ruby_active_record_demo.sqlite3", Dir.pwd)

ActiveRecord::Base.establish_connection(
  adapter: "sqlite3",
  database: db_path
)

unless ActiveRecord::Base.connection.table_exists?("users")
  ActiveRecord::Schema.define do
    create_table :users do |t|
      t.string :name, null: false
      t.string :email, null: false
      t.timestamps null: false
    end
  end
end

class User < ActiveRecord::Base
  self.table_name = "users"
end

view = ActionView::Base.empty
view.extend ActionView::Helpers::FormHelper
view.extend ActionView::Helpers::TagHelper
view.extend ActionView::Context
```

# Active Record
## Basic CRUD

```ruby
User.delete_all

created_user = User.create!(name: "Ada Lovelace", email: "ada@example.com")
puts "Created:"
puts created_user.attributes

read_user = User.find_by(email: "ada@example.com")
puts "Read:"
puts read_user.attributes

read_user.update!(name: "Ada Byron")
puts "Updated:"
puts read_user.reload.attributes

read_user.destroy!
puts "Destroyed: #{read_user.destroyed?}"
puts "Remaining users: #{User.count}"
```

```ruby
html = view.form_with(url: "/users", scope: :user) do |f|
  view.safe_join(
    [
      f.label(:name, "Name"),
      f.text_field(:name),
      f.label(:email, "Email"),
      f.email_field(:email),
      f.submit("Save")
    ],
    view.tag.br
  )
end

puts html
html
```
